```python
from collections import Counter


def extract_most_frequent_word(series):
    """
    Возвращает Series с самым частым словом из списков в исходной колонке.
    """
    def get_top_word(word_list):
        # Проверка на корректный тип и непустой список
        if not isinstance(word_list, list) or len(word_list) == 0:
            return np.nan
        # Counter.most_common(1) -> [('слово', count)]
        return Counter(word_list).most_common(1)[0][0]
        
    return series.apply(get_top_word)
```

```python
df['lable_silver_common_AI'] = extract_most_frequent_word(df["ALL_AI_LABELS"])
```

```python
import re
import pandas as pd


def normalize_spaces(text: str) -> str:
    text = re.sub(r"\s+", " ", text)
    return text.strip()


def remove_html(text: str) -> str:
    return re.sub(r"<.*?>", " ", text)


def mask_personal_data(text: str) -> str:
    # email
    text = re.sub(
        r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}\b",
        "<EMAIL>",
        text
    )

    # phone numbers, rough version
    text = re.sub(
        r"(\+7|8)?[\s\-()]?\d{3}[\s\-()]?\d{3}[\s\-()]?\d{2}[\s\-()]?\d{2}",
        "<PHONE>",
        text
    )

    # urls
    text = re.sub(
        r"(https?://\S+|www\.\S+)",
        "<URL>",
        text
    )

    return text


def reduce_repeated_chars(text: str) -> str:
    # "круууутооооо" -> "круутоо"
    # Не режем до одной буквы, чтобы сохранить эмоциональность
    return re.sub(r"(.)\1{3,}", r"\1\1", text)


def has_enough_cyrillic(text: str, threshold: float = 0.5) -> bool:
    letters = re.findall(r"[A-Za-zА-Яа-яЁё]", text)
    if not letters:
        return False

    cyrillic = re.findall(r"[А-Яа-яЁё]", text)
    return len(cyrillic) / len(letters) >= threshold


def clean_review_text(text: str) -> str:
    if not isinstance(text, str):
        return ""

    text = remove_html(text)
    text = mask_personal_data(text)
    text = reduce_repeated_chars(text)
    text = normalize_spaces(text)

    return text


def make_dedup_key(text: str) -> str:
    text = text.lower()
    text = re.sub(r"[^\wа-яё]+", " ", text, flags=re.IGNORECASE)
    text = re.sub(r"\s+", " ", text)
    return text.strip()
```

```python
df_test = pd.read_excel("data/app_reviews_ru_only_text_AI_LABELED_PREPARED.xlsx")

# Сохраняем оригинал
df_test["review_text_raw"] = df_test["text"].astype(str)

# Создаём очищенный текст
df_test["review_text_clean"] = df_test["review_text_raw"].apply(clean_review_text)

# Удаляем пустые и слишком короткие
df_test = df_test[df_test["review_text_clean"].str.len() >= 3]

# Фильтр по русскому языку
df_test = df_test[df_test["review_text_clean"].apply(has_enough_cyrillic)]

# Удаляем дубликаты
df_test["dedup_key"] = df_test["review_text_clean"].apply(make_dedup_key)
df_test = df_test.drop_duplicates(subset=["dedup_key"])

# Удаляем временную колонку
df_test = df_test.drop(columns=["dedup_key"])

# Добавляем поля для ручной разметки
df_test["label_gold"] = ""
df_test["summary_gold"] = ""
df_test["comment"] = ""


# df_test.to_csv("app_reviews_ru_for_annotation.csv", index=False, encoding="utf-8-sig")
df_test.to_excel("data/app_reviews_ru_AI_LABELED_for_annotation.xlsx", index=False)

print(f"Готово. Осталось отзывов: {len(df)}")
```
